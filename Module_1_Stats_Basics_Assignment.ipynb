Q1. What is the difference between descriptive statistics and inferential statistics? Explain with examples.

Ans:
**Descriptive Statistics**
* Definition:
Descriptive statistics refers to the methods used to summarize, organize, and present data in a meaningful way. It involves techniques that describe the main features of a dataset without making conclusions beyond the data.

  * Purpose: To provide a snapshot of the data through charts, tables, and numerical summaries.

* Common Techniques:

   * Measures of central tendency (mean, median, mode)
   * Measures of dispersion (range, variance, standard deviation)
   * Data visualization (bar charts, pie charts, histograms, box plots)

Example:
Suppose a teacher records the marks of 50 students in a mathematics test. Descriptive statistics might include:

The average (mean) mark is 72.

The highest mark is 95 and the lowest is 45.

A histogram shows how the marks are distributed.

In this case, descriptive statistics are used to summarize the marks for these 50 students only â€” no predictions or generalizations are made about other students or future tests.



**Inferential Statistics**
* Definition:
Inferential statistics involves methods for making predictions, estimates, or decisions about a larger population based on a sample of data. It helps determine the likelihood that a pattern in the data reflects the population or occurred by chance.

* Purpose:
To make inferences or generalizations beyond the immediate data.

* Common Techniques:
    * Hypothesis testing
    * Confidence intervals
    * Regression analysis
    * ANOVA (Analysis of Variance)
    * Chi-square tests

* Example:
A political analyst surveys 1,000 people in a country to determine the percentage of the population that supports a particular candidate. Based on the sample:
  * They estimate that 55% of the population supports the candidate, with a 95% confidence interval of Â±3%.
  * They perform a hypothesis test to check if the candidate has more than 50% support in the general population.
  * In this case, inferential statistics are used to draw conclusions about the entire population based on data from a sample.





Q2. What is sampling in statistics? Explain the differences between random
and stratified sampling.

Ans: Sampling is the process of selecting a subset (sample) of individuals from a larger population to collect data and make inferences about the entire population.

Because it is often impractical or too costly to collect data from every member of a population, researchers use sampling to gather data more efficiently.

* Purpose of Sampling
   * To save time and resources
   * To make predictions or generalizations about a population
   * To conduct statistical analysis using manageable data sizes


**Random Sampling**
  * Definition:
Random sampling is a method where every member of the population has an equal chance of being selected.

 * How it works:
You assign numbers to each member of the population and use a random number generator or lottery method to choose the sample.

* Key Features:
   * Completely unbiased selection
   * Every individual has an equal probability
   * Simple to understand and implement (especially with computers)

* Example:
If you want to survey 100 students out of 1,000 in a university, you could use a random number generator to select 100 student IDs.

* Advantages:
  * Reduces selection bias
  * Results are usually representative if the sample is large enough

* Disadvantages:
  * May not represent important sub-groups within the population (e.g., age, gender, location)


**Stratified Sampling**
* Definition:
Stratified sampling involves dividing the population into sub-groups (strata) based on shared characteristics, and then randomly sampling from each stratum.

* Strata could be based on:
  * Age
  * Gender
  * Income level
  * Education level, etc.

* How it works:
  * Divide the population into mutually exclusive groups (strata).
  * Perform random sampling within each group.
  * Combine the samples from all strata to form the final sample.

* Example:
 You want to survey 200 people about healthcare opinions. You know that 60% of the population is urban and 40% is rural. You divide your population into urban and rural groups, then randomly sample 120 people from the urban group and 80 from the rural group.

* Advantages:
  * Ensures representation of all key subgroups
  * Can lead to more accurate and reliable results

* Disadvantages:
  * More complex and time-consuming to implement
  * Requires detailed knowledge of population structure

Q3. Define mean, median, and mode. Explain why these measures of central
tendency are important.

Ans:   Measures of Central Tendency
Measures of central tendency are statistical tools used to describe the center or typical value of a dataset. The three most commonly used measures are:

* Mean

* Median

* Mode

They help to summarize data with a single value that represents the entire distribution.

* Mean (Average)
Definition:
The mean is the sum of all values in a dataset divided by the number of values.
* Strengths:
 * Takes all data values into account
 * Widely used in analysis and statistics

* Weakness:
 * Affected by extreme values (outliers)

* Median
Definition:
The median is the middle value when the data is arranged in ascending or descending order.

If there is an odd number of values, the median is the middle one.

If there is an even number of values, the median is the average of the two middle values.

* Example (Odd):
Data: 10, 15, 20, 25, 30
â†’ Median = 20 (middle value)

* Example (Even):
Data: 10, 20, 30, 40
â†’ Median = (20 + 30) Ã· 2 = 25

* Strengths:
  * Not affected by outliers or skewed data
  * Better representation when data is not symmetrically distributed

* Weakness:
  * Doesn't consider all values in the dataset

* Mode
Definition:
The mode is the value that appears most frequently in a dataset.

* Example:
Data: 3, 5, 7, 5, 9, 5, 10
â†’ Mode = 5 (appears 3 times)



* Strengths:
  * Useful for categorical data (e.g., most common brand, color, etc.)
  * Easy to identify

* Weakness:
  * May not represent the data well if multiple modes or no clear mode exist

Q4. Explain skewness and kurtosis. What does a positive skew imply about the
data?
Ans:
Skewness measures the asymmetry of a data distribution around its mean.
  * If data is symmetrical, the skewness is zero.
  * If data is asymmetrical, skewness can be positive or negative.

* What Does a Positive Skew Imply About the Data?
 * Most data values are clustered on the left (lower values).
 * There are a few large values (outliers) that stretch the tail to the right
 The mean is greater than the median.

* Often found in income data, real estate prices, or waiting times (many people have low values, few have very high).

* ðŸ“Œ Example:
Income of 10 people:
20k, 22k, 24k, 25k, 26k, 27k, 28k, 29k, 30k, 120k
â†’ Most values are around 20k-30k, but one person earns 120k â†’ Positive skew.

* Kurtosis
* Definition:
Kurtosis measures the "tailedness" or peakedness of a distribution â€” i.e., how much of the variance is due to extreme values (outliers).

* It tells us how heavy or light the tails of a distribution are compared to a normal distribution.

* Why Are Skewness and Kurtosis Important?
   * Help assess whether data is normally distributed â€” a key assumption in many statistical tests.
   * Detect outliers and understand the shape of the data.
   * Assist in choosing the right statistical methods (some tests assume normality).

```python
#Q6. Compute the covariance and correlation coefficient between the following two datasets provided as lists in Python:
# list_x = [10, 20, 30, 40, 50]
# list_y = [15, 25, 35, 45, 60]
import numpy as np

# Given data
list_x = [10, 20, 30, 40, 50]
list_y = [15, 25, 35, 45, 60]

# Convert to NumPy arrays
x = np.array(list_x)
y = np.array(list_y)

# Calculate covariance matrix
cov_matrix = np.cov(x, y, bias=False)  # bias=False gives sample covariance
covariance = cov_matrix[0, 1]

# Calculate correlation coefficient
correlation = np.corrcoef(x, y)[0, 1]

# Print results
print("Covariance:", covariance)
print("Correlation Coefficient (r):", correlation)
```

```python
#Q7. Write a Python script to draw a boxplot for the following numeric list and identify its outliers. Explain the result:
# data = [12, 14, 14, 15, 18, 19, 19, 21, 22, 22, 23, 23, 24, 26, 29, 35]
import matplotlib.pyplot as plt
import numpy as np

# Given data
data = [12, 14, 14, 15, 18, 19, 19, 21, 22, 22, 23, 23, 24, 26, 29, 35]

# Create the boxplot
plt.boxplot(data, vert=False)
plt.title("Boxplot of Given Data")
plt.xlabel("Value")
plt.grid(True)
plt.show()

# Outlier detection using IQR method
Q1 = np.percentile(data, 25)
Q3 = np.percentile(data, 75)
IQR = Q3 - Q1

# Define outlier boundaries
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Find outliers
outliers = [x for x in data if x < lower_bound or x > upper_bound]

# Print results
print(f"Q1 (25th percentile): {Q1}")
print(f"Q3 (75th percentile): {Q3}")
print(f"IQR (Q3 - Q1): {IQR}")
print(f"Lower Bound: {lower_bound}")
print(f"Upper Bound: {upper_bound}")
print(f"Outliers: {outliers}")
```

```python
#Q8. You are working as a data analyst in an e-commerce company. The marketing team wants to know if there is a relationship between advertising spend and daily sales.
#  Explain how you would use covariance and correlation to explore this relationship.
# Write Python code to compute the correlation between the two lists: advertising_spend = [200, 250, 300, 400, 500] daily_sales = [2200, 2450, 2750, 3200, 4000]


#  Using Covariance and Correlation to Explore Relationships
# 1. Covariance
# Definition: Measures the direction of the relationship between two variables.
# If covariance is:
# Positive: As advertising spend increases, daily sales tend to increase.
# Negative: As advertising spend increases, daily sales tend to decrease.
# Limitation: It doesn't tell us the strength of the relationship and is affected by the scale of the variables.

# 2. Correlation (Pearsonâ€™s r)
# Definition: Measures both the direction and strength of a linear relationship between two variables.
# Range: -1 to 1
# +1: Perfect positive correlation
# 0: No correlation
# âˆ’1: Perfect negative correlation
# Interpretation:
# A value close to 1 means a strong positive linear relationship between advertising spend and daily sales.
# Unlike covariance, correlation is scale-independent.

import numpy as np

# Given data
advertising_spend = [200, 250, 300, 400, 500]
daily_sales = [2200, 2450, 2750, 3200, 4000]

# Convert to numpy arrays
x = np.array(advertising_spend)
y = np.array(daily_sales)

# Calculate covariance matrix
cov_matrix = np.cov(x, y, bias=False)
covariance = cov_matrix[0, 1]

# Calculate correlation coefficient
correlation = np.corrcoef(x, y)[0, 1]

# Print results
print("Covariance:", covariance)
print("Correlation Coefficient:", correlation)
```

```python
#Q9. Your team has collected customer satisfaction survey data on a scale of 1-10 and wants to understand its distribution before launching a new product.
# Explain which summary statistics and visualizations (e.g. mean, standard deviation, histogram) youâ€™d use.
# Write Python code to create a histogram using Matplotlib for the survey data: survey_scores = [7, 8, 5, 9, 6, 7, 8, 9, 10, 4, 7, 6, 9, 8, 7]

# To understand the distribution of customer satisfaction survey scores on a scale of 1 to 10, we would begin by analyzing a few key summary statistics. The mean provides the average score, indicating the general level of satisfaction among customers. If the mean is high (close to 10), it suggests that customers are generally satisfied. The median, or middle value when scores are ordered, is useful to assess central tendency, especially if the data contains outliers or is skewed. The mode reveals the most frequently occurring satisfaction score, which can point to common customer sentiment. To understand the variability in responses, we use the standard deviationâ€”a higher value would indicate that opinions are more spread out, while a lower standard deviation means most customers rated satisfaction similarly. Additionally, looking at the minimum and maximum values helps understand the full range of customer opinions.

# For visualization, a histogram is particularly helpful in showing how scores are distributed across the scale. It can reveal whether the scores are clustered in a specific range, symmetrically distributed, or skewed toward higher or lower values. A boxplot can also be helpful to quickly identify the median, quartiles, and potential outliers in the data. However, a histogram remains the go-to option when trying to assess the overall shape and spread of the satisfaction scores.

import matplotlib.pyplot as plt
import statistics

# Survey data
survey_scores = [7, 8, 5, 9, 6, 7, 8, 9, 10, 4, 7, 6, 9, 8, 7]

# Calculate summary statistics
mean_score = statistics.mean(survey_scores)
median_score = statistics.median(survey_scores)
mode_score = statistics.mode(survey_scores)
std_dev = statistics.stdev(survey_scores)

# Print summary
print(f"Mean: {mean_score}")
print(f"Median: {median_score}")
print(f"Mode: {mode_score}")
print(f"Standard Deviation: {std_dev}")

# Create histogram
plt.hist(survey_scores, bins=7, edgecolor='black', color='skyblue')
plt.title('Customer Satisfaction Score Distribution')
plt.xlabel('Satisfaction Score (1â€“10)')
plt.ylabel('Number of Responses')
plt.axvline(mean_score, color='red', linestyle='dashed', linewidth=1, label=f'Mean: {mean_score:.2f}')
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
